# **Import Library**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
url = "/content/drive/MyDrive/Machine Learning/Minggu 4/diamonds.csv"

# Membaca file CSV dan membuat DataFrame
df = pd.read_csv(url)

# Menampilkan beberapa baris pertama dari DataFrame
print(df.head())
```

# **Feature Statistic**

```python
# Menampilkan statistik fitur
feature_statistics = df.describe()
print(feature_statistics)
```

# **Box Plot**

```python
df.info()
```

```python
# Menampilkan boxplot untuk setiap fitur numerik
numerical_features = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Membuat boxplot untuk setiap fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
```

# **Scatter Plot**

```python
# Menampilkan scatter plot untuk semua fitur numerik
scatter_matrix(df, figsize=(12, 12), alpha=0.5,)
plt.suptitle('Scatter Plot untuk Fitur Numerik', fontsize=16)
plt.show()
```

# **Distribution Plot**

```python
# Menampilkan distribution plot untuk setiap fitur numerik
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Distribution Plot untuk Fitur Numerik', fontsize=16)
plt.show()
```

# **Violin Plot**

```python
# Menampilkan violin plot untuk semua fitur numerik
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x=df[feature], inner='quartile')
    plt.title(f'Violin Plot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

# **Heatmap**

```python
# Menghitung korelasi antara fitur-fitur numerik
correlation_matrix = df.corr()

# Menampilkan heatmap dari korelasi antara fitur-fitur numerik
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Heatmap dari Korelasi Fitur Numerik', fontsize=16)
plt.show()
```
